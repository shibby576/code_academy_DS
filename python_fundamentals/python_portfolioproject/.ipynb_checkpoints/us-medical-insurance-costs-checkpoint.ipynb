### Scope
-look at average costs for smoker vs non and bmi levels

-What region has the highest bmi?

-What is the biggest driver of cost?

```python
#Organize csv into lists of attributes
age=[]
sex=[]
bmi=[]
children=[]
smoker=[]
region=[]
charges=[]
with open('insurance.csv',newline='') as ins_file:
    ins_csv = csv.DictReader(ins_file)
    for row in ins_csv:
        age.append(row['age'])
        sex.append(row['sex'])
        bmi.append(row['bmi'])
        children.append(row['children'])
        region.append(row['region'])
        smoker.append(row['smoker'])
        charges.append(row['charges'])

```

```python
#avg cost of smoker vs non smoker

combined_cost=list(zip(smoker,charges))
total_smoker_cost=0
total_nonsmoker_cost=0
smoke_counter=0
non_smoke_counter=0

for i in combined_cost:
    if i[0] =='yes':
        total_smoker_cost +=float(i[1])
        smoke_counter+=1
    elif i[0]=='no':
        total_nonsmoker_cost +=float(i[1])
        non_smoke_counter +=1
        

average_smoker_cost = total_smoker_cost / smoke_counter
avg_nonsmoker_cost = total_nonsmoker_cost / non_smoke_counter

print(average_smoker_cost)
print(avg_nonsmoker_cost)
```
